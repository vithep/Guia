# Visualização de Dados

#### Importação das Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Importação de Dados

```python
df = pd.read_csv('https://raw.githubusercontent.com/nickssonarrais/datascience/main/storytelling/resultado_final.csv')
df.head()
```

## 1. Gráfico de Pizza

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(12,9))
ax[0].pie(df.Antes,labels=df.Status,autopct='%1.2f%%',shadow=True,startangle=90);
ax[1].pie(df.Depois,labels=df.Status,autopct='%1.2f%%',shadow=True,startangle=90);
```

## 2. Gráfico de Barras

```python
df.plot(kind='bar')
```

```python
labels = list(df.Status)
labels= ['Entediado',
         'Não muito\ninteressado',
         'Bem',
         'Um pouco\ninteressado',
         'Muito\nInteressado']
```

```python
fig,ax = plt.subplots(nrows=1,ncols=1)
df.plot(kind='bar',
        title='Como você se sente em relação à Ciência',
        figsize=(9,5),
        rot=30,
        ylabel='Porcentagem',
        xlabel='',
        ax=ax);

ax.set_xticklabels(labels);
```

#### Plotando as barras separadamente e modificando as cores

```python
cinza_claro = "#CFCDCE"
cinza_medio = "#A2A1A3"
cinza_escuro = "#787878"
azul_claro = "#8db7eb"
azul_escuro = "#253760"
color_antes = [cinza_claro, cinza_claro, cinza_claro, cinza_medio, cinza_escuro]
color_depois = [azul_claro, azul_claro, azul_claro, azul_escuro, azul_escuro]
```

```python
#criando figura e eixo para gráfico de barras
fig, ax = plt.subplots(figsize=(8,4))

#criando um vetor para cada elemento Status
x = np.array(range(1,df.shape[0]+1))

#plotar barras antes e depois separadamente
width_bar=0.4
barras_antes = ax.bar(x-width_bar/2,df.Antes,width_bar,color=color_antes)
barras_depois = ax.bar(x+width_bar/2,df.Depois,width_bar,color=color_depois)
```

#### Remover elementos desnecessários

```python
# Criando figura e eixo para gráfico de barras
fig,ax = plt.subplots(figsize=(8,4))

# Plotar antes e depois separadamente
width_bar=0.4
barras_antes=ax.bar(x-width_bar/2,df.Antes,width_bar,color=color_antes)
barras_depois=ax.bar(x+width_bar/2,df.Depois,width_bar,color=color_depois)

# Remover grids e eixos 
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remover eixo y
plt.gca().axes.get_yaxis().set_visible(False)

# Remover eixo X
#plt.gca().axes.get_xaxis().set_visible(False)

# Remover todos os eixos
#ax.set_axis_off()
```

#### Adicionando Elementos

```python
# Criando figura e eixo para gráfico de barras
fig,ax = plt.subplots(figsize=(8,4))

# Plotar antes e depois separadamente
width_bar=0.4
barras_antes=ax.bar(x-width_bar/2,df.Antes,width_bar,color=color_antes)
barras_depois=ax.bar(x+width_bar/2,df.Depois,width_bar,color=color_depois)

# Remover grids e eixos 
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remover eixo y
plt.gca().axes.get_yaxis().set_visible(False)

## Adicionando Elementos

# Label do eixo
ax.set_xticks(x)
ax.set_xticklabels(labels);

x_text_antes = 0.2
# Adicionando um título
plt.text(x=x_text_antes,
        y=52,
        s='Como você se sente em relação à Ciência',
        fontsize=16,
        color=cinza_escuro,
        weight='bold');

# Resultado da Campanha
plt.text(x=x_text_antes,
        y=49,
        s="$\\bf{Antes}$ do programa\n"
        "a maioria das crianças\n"
        "se sentia BEM em\n"
        "relação à ciência.",
        verticalalignment="top",
        fontsize=12,
        color=cinza_claro);

# Resultado da Campanha Depois
plt.text(x=5.5,
        y=39,
        s="$\\bf{DEPOIS}$\n"
        "do programa\n"
        "mais crianças tinham\n"
        "algum interesse e\n"
        "muito interesse em ciência.",
        verticalalignment="top",
        fontsize=12,
        color=azul_escuro);
```

#### Adicionando Porcentagem 

```python
#criar função para dar labels automaticamente
def set_label_y_in_bar(bar):
    for b in bar:
        height_bar = b.get_height()
        ax.annotate(
                    '{}%'.format(height_bar),
                    xy=(b.get_x() + b.get_width()/2,  height_bar),
                    xytext=(0, -20),  # offset
                    textcoords="offset points",
                    ha='center',
                    va='bottom',
                    fontsize=10,
                    color="white",
                    )
```

```python
# Criando figura e eixo para gráfico de barras
fig,ax = plt.subplots(figsize=(8,4))

# Plotar antes e depois separadamente
width_bar=0.4
barras_antes=ax.bar(x-width_bar/2,df.Antes,width_bar,color=color_antes)
barras_depois=ax.bar(x+width_bar/2,df.Depois,width_bar,color=color_depois)

# Remover grids e eixos 
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remover eixo y
plt.gca().axes.get_yaxis().set_visible(False)

## Adicionando Elementos

# Label do eixo
ax.set_xticks(x)
ax.set_xticklabels(labels);

x_text_antes = 0.2
# Adicionando um título
plt.text(x=x_text_antes,
        y=52,
        s='Como você se sente em relação à Ciência',
        fontsize=16,
        color=cinza_escuro,
        weight='bold');

# Resultado da Campanha
plt.text(x=x_text_antes,
        y=49,
        s="$\\bf{Antes}$ do programa\n"
        "a maioria das crianças\n"
        "se sentia BEM em\n"
        "relação à ciência.",
        verticalalignment="top",
        fontsize=12,
        color=cinza_claro);

# Resultado da Campanha Depois
plt.text(x=5.5,
        y=39,
        s="$\\bf{DEPOIS}$\n"
        "do programa\n"
        "mais crianças tinham\n"
        "algum interesse e\n"
        "muito interesse em ciência.",
        verticalalignment="top",
        fontsize=12,
        color=azul_escuro);

# Colocando a Porcentagem
set_label_y_in_bar(barras_antes)
set_label_y_in_bar(barras_depois)

# plotar o gráfico
fig.tight_layout()
plt.show()
```
