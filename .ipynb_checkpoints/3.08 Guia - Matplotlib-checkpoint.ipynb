# 3.8 Guia - Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gerando os numeros com o numpy
dados_x = np.linspace(0,10,10)
dados_y = dados_x*dados_x
```

```python
# configurar a visualização
#%matplotlib inline
%config InlineBackend.figure_format = 'svg'
#configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'
```

```python
# Gráfico Linha com 2 eixos e ajustando os parâmetros
# Ajustando a figura
plt.figure(figsize=(10,5), 
           facecolor='gray') # Ajustando a cor

# Adicionando eixos
ax = plt.axes()
ax.set_facecolor('white') #Cor de fundo do gráfico

# Remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Criando o gráfico
plt.plot(
        dados_x,dados_y,
        color='magenta', # Cor da linha
        linewidth=2, # Espessura da Linha
        markersize=5, # Tamanho do Marcador
        marker='o', # Tipo de Marcador
        linestyle='dashed', # Tipo de linha
        markeredgecolor='black' # Cor linha do Marcador
)

# Definindo label do eixo X
plt.xlabel(
        'Eixo X', # Nome do Label
        fontweight='bold', # Estilo
        fontsize='large', # Tamanho da Fonte
        fontfamily='fantasy', # Tipo de Fonte
        color = 'Blue' # Cor da Fonte
)

# Definindo label do eixo Y
plt.ylabel('Eixo Y')

# Definindo um Titulo
plt.text(
        x = 0,
        y = 110, # Coordenada para o Titulo
        s = 'Gráfico Simples', # Texto
        fontsize=16, # Tamanho do Texto
        fontfamily='fantasy', # Tipo da Fonte
        color='white', # Cor do Texto
        weight='bold' #
)

# Definindo as Legendas
plt.legend(
        'Número Aleatórios', # Nome da Legenda
        loc='upper center', # Posição da Legenda
        bbox_to_anchor=(1,0.,0.0,0.2), # Posição da legenda
        fontsize='large', # Tamanho do Texto
        shadow=True, # Sombra na Legenda
        borderpad=1 # Tamanho da borda
        );
# % da Posição na esquerda
# % da Posição inferior
# % da Largura
# % da Altura
```

```python
# Plotando varios graficos em uma grade
Linhas = 2
Colunas = 2

# Ajustando a figura
plt.figure(figsize=(10,8), 
           facecolor='gray') # Ajustando a cor

# Adicionando eixos
ax = plt.axes()
ax.set_facecolor('white') #Cor de fundo do gráfico

# Primeiro Grafico
plt.subplot(Linhas,Colunas,1)
plt.plot(dados_x,'r--')
plt.title('Grafico 1')
plt.legend(
        'x', # Nome da Legenda
        loc='upper center', # Posição da Legenda
        bbox_to_anchor=(1,0.,0.0,0.2), # Posição da legenda
        fontsize='large', # Tamanho do Texto
        shadow=True, # Sombra na Legenda
        borderpad=1 # Tamanho da borda
        );

# Segundo Grafico
plt.subplot(Linhas,Colunas,2)
plt.plot(dados_y,'g-')
plt.title('Grafico 2')
plt.legend(
        'y', # Nome da Legenda
        loc='upper center', # Posição da Legenda
        bbox_to_anchor=(1,0.,0.0,0.2), # Posição da legenda
        fontsize='large', # Tamanho do Texto
        shadow=True, # Sombra na Legenda
        borderpad=1 # Tamanho da borda
        );

# Terceiro Grafico
plt.subplot(Linhas,Colunas,3)
plt.plot(dados_y,'y-')
plt.title('Grafico 3')
plt.legend(
        'y', # Nome da Legenda
        loc='upper center', # Posição da Legenda
        bbox_to_anchor=(1,0.,0.0,0.2), # Posição da legenda
        fontsize='large', # Tamanho do Texto
        shadow=True, # Sombra na Legenda
        borderpad=1 # Tamanho da borda
        );

# Quarto Grafico
plt.subplot(Linhas,Colunas,4)
plt.plot(dados_x,'b--')
plt.plot(dados_y,'p--')
plt.title('Grafico 4')
plt.legend(
        'xy', # Nome da Legenda
        loc='upper center', # Posição da Legenda
        bbox_to_anchor=(0,0.45,0.3,0.5), # Posição da legenda
        fontsize='large', # Tamanho do Texto
        shadow=True, # Sombra na Legenda
        borderpad=1 # Tamanho da borda
        );

# Definindo um Titulo
plt.text(
        x = -5,
        y = 260, # Coordenada para o Titulo
        s = 'Gráficos Separados', # Texto
        fontsize=16, # Tamanho do Texto
        fontfamily='fantasy', # Tipo da Fonte
        color='Black', # Cor do Texto
        weight='bold' #
);
```

```python
# Plotando gráficos de diferentes tamanhos

# Criando a figura
Figura = plt.figure(figsize=(10,6), 
           facecolor='gray')

# Definindo o tamanho do grafico
Eixo_1 = Figura.add_axes([0.1,0.1,0.8,0.8])

# Definindo o tamanho do grafico
Eixo_2 = Figura.add_axes([0.2,0.5,0.3,0.3])

# Plotando os graficos
Eixo_1.plot(dados_x,dados_y,'r--')
Eixo_2.plot(dados_x,dados_y,'y');
```

```python
# Grafico de Barras

# Criando figura
Figura = plt.figure()

# Definindo o Tamanho
Grafico = Figura.add_axes([0,0,0.7,0.7])

# Definindo os rotulos
Rotulos = ['A','B','C','D','E','F','G','H','I','J']

# Gerando o gráfico
Grafico.bar(Rotulos,dados_x);
```

```python
# Grafico de Barras com mais eixos

# Criando figura
Figura = plt.figure()

# Definindo o Tamanho
Grafico = Figura.add_axes([0,0,0.7,0.7])

# Definindo os rotulos
Rotulos = ['A','B','C','D','E','F','G','H','I','J']

# Gerando o gráfico
Grafico.bar(Rotulos,dados_x,color='r')
Grafico.bar(Rotulos,dados_y*0.1,color='b');
```
