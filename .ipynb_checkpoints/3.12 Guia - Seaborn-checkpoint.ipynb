# 3.12 Guia - Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_dados=sns.load_dataset('tips')
base_dados.head()
```

```python
# Mudando o nome
base_dados.rename(columns={
    'total_bill' : 'Total Conta',
    'tip' : 'Gorjeta',
    'sex' : 'Sexo',
    'smoker' : 'Fumante',
    'day' : 'Dia da Semana',
    'time' : 'Periodo',
    'size' : 'Pessoas na mesa'}, inplace=True
)
```

```python
base_dados.columns
```

```python
# configurar a visualização
#%matplotlib inline
%config InlineBackend.figure_format = 'svg'
#configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'
```

```python
# Gráfico Linha com 2 eixos e ajustando os parâmetros

# Criando o gráfico
sns.relplot(
        x='Total Conta',y='Gorjeta',
        data=base_dados,
        color='magenta', # Cor 
        marker='o' # Tipo de Marcador
)
sns.set_theme(style='darkgrid')


# Definindo um Titulo
plt.text(
        x = 0,
        y = 11, # Coordenada para o Titulo
        s = 'Conta x Gorjeta', # Texto
        fontsize=16, # Tamanho do Texto
        fontfamily='fantasy', # Tipo da Fonte
        color='black', # Cor do Texto
        weight='bold' #
)

# Definindo as Legendas
plt.legend(
        'P', # Nome da Legenda
        loc='upper center', # Posição da Legenda
        bbox_to_anchor=(1,0.,0.0,0.2), # Posição da legenda
        fontsize='large', # Tamanho do Texto
        shadow=True, # Sombra na Legenda
        borderpad=1 # Tamanho da borda
        );
```

```python
# Gráfico Linha com 2 eixos e ajustando os parâmetros

# Criando o gráfico
sns.relplot(
        x='Total Conta',y='Gorjeta',
        data=base_dados,
        color='magenta', # Cor 
        marker='o',
        hue = 'Sexo' # Tipo de Marcador
)
sns.set_theme(style='darkgrid')


# Definindo um Titulo
plt.text(
        x = 0,
        y = 11, # Coordenada para o Titulo
        s = 'Conta x Gorjeta', # Texto
        fontsize=16, # Tamanho do Texto
        fontfamily='fantasy', # Tipo da Fonte
        color='black', # Cor do Texto
        weight='bold' #
)
```
