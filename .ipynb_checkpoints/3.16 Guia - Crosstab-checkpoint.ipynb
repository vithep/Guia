```python
import pandas as pd
```

```python
df = pd.read_excel("D:\\PythonProjects\\survey_1.xlsx")
df
```

```python
#using crosstab

#In statistics, a contingency table (also known as a cross tabulation or crosstab) 
#is a type of table in a matrix format
# It displays the (multivariate) frequency distribution of the variables.
# They provide a basic picture of the interrelation between two variables and can help find interactions between them.

#Nationality on X-axis (i.e rows) , Handedness on Y axis (i.e columns). --> These are variables 
#It tells frequency of how elements are occuring --> count of handedness


pd.crosstab(df.Nationality, df.Handedness) 
```

```python
#Gender on X-axis (rows) and Handedness on Y-axis (columns) --> variables
#Press Shift + Tab --> to open documentation about methods 
pd.crosstab(df.Gender, df.Handedness)
```

```python
#Calculate total rowwise and coolumwise, use of margins with crosstab
pd.crosstab(df.Gender, df.Handedness, margins=True)
```

```python
#Adding nationality at column level
pd.crosstab(df.Gender, [df.Handedness, df.Nationality], margins=True)
```

```python
#Adding nationality at row level
#x-axis (row level) --> Nationality and Gender
pd.crosstab([df.Nationality,df.Gender], [df.Handedness], margins=True)
```

```python
#Adding percentages
#shown All... This helps to understand percentages shown in next step
pd.crosstab([df.Gender], [df.Handedness], margins=True)
```

```python
#Adding percentages
#use normalize parameter with index to normalize at row level. This gives percentage
pd.crosstab([df.Gender], [df.Handedness], normalize='index')
```
