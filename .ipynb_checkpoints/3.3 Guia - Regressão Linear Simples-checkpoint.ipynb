# 3.3 Guia - Regressão Linear Simples

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Base de dados

Metragem = [40,45,50,55,60,62,65,70,80,90,92,100,110,120,150]
Valor    = [200,280,310,350,390,410,450,490,550,620,670,700,750,810,989]

# Organizando em um Dicionario 
Dicionario = {
        'Metragem': Metragem,
        'Valor Imovel': Valor
}

DataFrame = pd.DataFrame(data=Dicionario)
DataFrame.head()
```

```python
# configurar a visualização
#%matplotlib inline
%config InlineBackend.figure_format = 'svg'
#configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'
```

```python
# Plotando os Dados

fig,ax = plt.subplots(figsize=(10,5))

ax.scatter(DataFrame['Metragem'].values,DataFrame['Valor Imovel'].values)
plt.xlabel('Preço do Apartamento')
plt.ylabel('Valor do Imovel')

# Remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Adicionando um título
plt.text(x=40,
        y=1100,
        s='Preço do Apartamento',
        fontsize=16,
        color='grey',
        weight='bold');
```

```python
# Verificando a Correlação
Eixo_x = DataFrame.iloc[:,0].values
Eixo_y = DataFrame.iloc[:,1].values

Correlação = np.corrcoef(Eixo_x,Eixo_y)
Correlação
```

```python
# Converter o Eixo X para o formato de Matriz
Eixo_x = Eixo_x.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
# Variavel com os Calculos
Regressor = LinearRegression()
# Dados Treino
Regressor.fit(Eixo_x,Eixo_y)
# Identificando o Coeficiente e a Constante
B1 = Regressor.coef_
B0 = Regressor.intercept_
# Mostrando os valores
print('B0 =', B0, '\n','B1 =', B1[0])
```

```python
# Calculando o Score
Score = Regressor.score(Eixo_x,Eixo_y)
Score
```

```python
# Plotando os Dados

fig,ax = plt.subplots(figsize=(10,5))

ax.scatter(DataFrame['Metragem'].values,DataFrame['Valor Imovel'].values)
ax.plot(Eixo_x, Regressor.predict(Eixo_x),color='red')
plt.xlabel('Preço do Apartamento')
plt.ylabel('Valor do Imovel')

# Remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Adicionando um título
plt.text(x=40,
        y=1100,
        s='Preço do Apartamento',
        fontsize=16,
        color='grey',
        weight='bold');

# Adicionando Legenda
plt.text(x=160,
        y=950,
        s='Regressão',
        fontsize=7,
        color='red',
        weight='bold');

plt.text(x=160,
        y=920,
        s='Dados',
        fontsize=7,
        color='blue',
        weight='bold');
```

```python
# Fazendo uma Previsão

# Qual o tamanho do Apartamento?
Qual_tamanho_apartamento = [[50]]

# Prevendo quanto custará o apartamento?
Previsão = Regressor.predict(Qual_tamanho_apartamento)

print('Um apartamento de: ',
     Qual_tamanho_apartamento[0][0], 'metros')
print('Usando o modelo, custaria: R$', round(Previsão[0],2))
```
