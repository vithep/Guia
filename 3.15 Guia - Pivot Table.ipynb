```python
%config IPCompleter.greedy=True	           #enable intellisense 
import pandas as pd
import numpy as np
```

```python
#pivot allows to reshape or transform a dataframe
df = pd.read_csv('D:\\PythonProjects\\multipleCities_2.csv')
df
```

```python
#using pivot
#set row(i.e x-axis) and columns (y-axis)
#we set Date as x-axis and city as columns (y-axis)
df.pivot(index="Date", columns="city")
```

```python
#using pivot
#only particular values to be shown (e.g. windspeed values only)
#set values="windspeed"
df.pivot(index="Date", columns="city", values="windspeed")
```

```python
#using pivot
#set row(i.e x-axis) and columns (y-axis)
#we set city as x-axis and Date as columns (y-axis)
df.pivot(index="city", columns="Date")
```

```python
#Using Pivot Table
df = pd.read_csv('D:\\PythonProjects\\multipleCities_3.csv')  #file has 2 readings of each day (like morning temp and evening temp)
df
```

```python
#Using Pivot Table
#Aggregation can be performed by using pivot table
#It calculates average of temperature and windspeed for 2 readings of each day (say morning temp and evening temp)
df.pivot_table(index="city",columns="Date")
```

```python
#Using Pivot Table
#Defining custom Aggregation function
#provide aggfunc argument
#aggfunc argument is numpy function
df.pivot_table(index="city",columns="Date", aggfunc=np.count_nonzero)
```

```python
#Using Pivot Table
#Defining custom Aggregation function
#provide aggfunc argument
#aggfunc argument is numpy function
df.pivot_table(index="city",columns="Date", aggfunc=np.mean)
```

```python
#Using Pivot Table
#Defining custom Aggregation function
#provide aggfunc argument
#aggfunc argument is numpy function
#More about numpy functions --> https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.math.html
#https://pandas.pydata.org/pandas-docs/stable/generated/pandas.pivot_table.html
df.pivot_table(index="city",columns="Date", aggfunc=np.sum)
```

```python
#using Pivot Table
#set margins .. 
#margins takes average of rows and columns and mentions it in All row and All column
df.pivot_table(index="city",columns="Date", margins=True)
```

```python
#using Grouper
df = pd.read_csv('D:\\PythonProjects\\OneCity_1.csv')
df
```

```python
#Using grouper
#First convert date which is string object to datetime series. This is extremely important step 
#otherwise Grouper will fail (in next step) as Grouper requires date as TimeStamp (type)

df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
type(df['Date'][0])   #type is now TimeStamp
```

```python
#using Grouper
#https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.Grouper.html
#http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases
#Aggregrate based on temperature , 
#using month (M) for frequency in Grouper
df.pivot_table(index=pd.Grouper(freq='M', key='Date'), columns='city') #shows average temperature and windspeed for each month
```
