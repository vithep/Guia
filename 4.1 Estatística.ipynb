# 4.1 Estatística

```python
from scipy.stats import norm
from scipy.stats import uniform
from scipy.stats import gamma
from scipy.stats import expon
from scipy.stats import randint  
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# configurar a visualização
#%matplotlib inline
%config InlineBackend.figure_format = 'svg'
#configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'
```

## Distribuição Normal

```python
normal = norm.rvs(size=1000,loc=0,scale=1)

ax = sns.distplot(normal,
                  bins=100,
                  kde=True,
                  color='Green')
ax.set(xlabel='Normal Distribution', ylabel='Frequency');

ax.text(
        x=-1.75,
        y=0.65,
        s='Distribuição Normal',
        fontsize=15);
```

## Distribuição Uniforme

```python
n = 1000
start = 10
width = 20
uniform_data = uniform.rvs(size=n, loc = start, scale=width)
```

```python
ax = sns.distplot(uniform_data,
                  bins=100,
                  kde=True,
                  color='red')
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')

ax.text(
        x=11,
        y=0.09,
        s='Distribuição Uniforme',
        fontsize=15);
```

## Distribuição Gamma

```python
gamma_data = gamma.rvs(a=5, size=10000)

ax = sns.distplot(gamma_data,
                  kde=True,
                  bins=100,
                  color='green')
ax.set(xlabel='Gamma Distribution', ylabel='Frequency')

ax.text(
        x=3.5,
        y=0.225,
        s='Distribuição Gamma',
        fontsize=15);
```

## Distribuição Exponencial

```python
data_expon = expon.rvs(scale=1,loc=0,size=1000)

ax = sns.distplot(data_expon,
                  kde=True,
                  bins=100,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')

ax.text(
        x=1,
        y=1.2,
        s='Distribuição Exponencial',
        fontsize=15);
```

## Distribuição Discreta

```python
numargs = randint .numargs  
a, b = 0.2, 0.8
rv = randint (a, b)  
    
print ("RV : \n", rv)     

distribution = np.linspace(0, np.minimum(rv.dist.b, 2))  
print("Distribution : \n", distribution)  
     
plot = plt.plot(distribution, rv.ppf(distribution))  
```

```python
x = np.linspace(0, 5, 100)  
     
# Varying positional arguments  
y1 = randint.ppf(x, a, b)  
y2 = randint.pmf(x, a, b)  
plt.plot(x, y1, "*", x, y2, "r--")
```
